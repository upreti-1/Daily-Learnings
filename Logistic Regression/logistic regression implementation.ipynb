# Logistic Regression Implementation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import make_classification
```

```python
# creating dataset
X, y = make_classification(n_samples= 1000, n_features= 10, n_classes= 2, random_state= 42)
```

```python
X
```

```python
y
```

```python
pd.DataFrame(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split (X, y, test_size= 0.30, random_state= 42)
```

```python
from sklearn.linear_model import LogisticRegression
logistic = LogisticRegression()
```

```python
logistic.fit(X_train, y_train)
```

```python
y_pred = logistic.predict(X_test)
print(y_pred)
```

```python
logistic.predict_proba(X_test)
```

```python
# performance matrics
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
score = accuracy_score(y_test, y_pred)
print('Confusion matrics: \n', score)
cm = confusion_matrix(y_test, y_pred)
print('Confusion matrics: \n', cm)
print(classification_report(y_test, y_pred))
```

# Hyperparameter Tuning And Cross Validation

# Grid SearchCV

```python
model = LogisticRegression()
penalty = ['l1', 'l2', 'elasticnet']
c_values = [100, 10, 0.1, 1]
solver = ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']
```

```python
params = dict(penalty = penalty, C = c_values, solver = solver)
```

```python
from sklearn.model_selection import StratifiedKFold
cv = StratifiedKFold()
```

```python
# GridSearchCV
# it findouts the best parameter for our model out of the upper given parameters
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(estimator= model, param_grid= params, scoring = 'accuracy', cv = cv, n_jobs= -1)
```

```python
grid
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred = grid.predict(X_test)
```

```python
score = accuracy_score(y_test, y_pred)
print('Confusion matrics: \n', score)
cm = confusion_matrix(y_test, y_pred)
print('Confusion matrics: \n', cm)
print(classification_report(y_test, y_pred))
```

# Randomized Search CV

# Logistic Regression for Imbalance Dataset

```python
from collections import Counter
from sklearn.datasets import make_classification
```

```python
X,y = make_classification(n_samples= 1000, n_features= 2, n_clusters_per_class= 1, n_redundant= 0, weights= [0.99], random_state= 42)
```

```python
X
```

```python
y
```

```python
Counter(y)
```

```python
import seaborn as sns
sns.scatterplot(x = pd.DataFrame(X)[0], y = pd.DataFrame(X)[1], hue = y)
```

```python
X_train, X_test, y_train, y_test = train_test_split (X, y, test_size= 0.25, random_state= 42)
```

```python
# Hyperparameter Tuning
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()


c_values = [100, 10, 0.1, 1]
solver = ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']
class_weight = [{0:w, 1:y} for w in [1, 10, 50, 100] for y in [1, 10, 50, 100]]
```

```python
params = dict(C = c_values, solver = solver, class_weight = class_weight)
```

```python
params
```

```python
from sklearn.model_selection import GridSearchCV, StratifiedKFold

# Ensure X_train and X_test have the SAME number of columns
print(f"Train shape: {X_train.shape}") # Should be (n, 10)
print(f"Test shape: {X_test.shape}")   # Should be (m, 10)

cv = StratifiedKFold(n_splits=5)
grid = GridSearchCV(estimator=model, param_grid=params, scoring='accuracy', cv=cv)

# 1. You MUST fit the grid search first
grid.fit(X_train, y_train)

# 2. Now you can predict
y_pred = grid.predict(X_test)
```

```python
score = accuracy_score(y_test, y_pred)
print('Confusion matrics: \n', score)
cm = confusion_matrix(y_test, y_pred)
print('Confusion matrics: \n', cm)
print(classification_report(y_test, y_pred))
```
