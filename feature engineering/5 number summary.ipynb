```python
import numpy as np
```

```python
list_marks = [45,32,34,45,99,98,67,74,43,97,34,32]
min, Q1, median, Q3, max = np.quantile(list_marks, [0, 0.25, 0.5, 0.75, 1])
```

```python
IQR = Q3 - Q1
print(IQR)
```

```python
lower_fence = Q1 - 1.5 * IQR
higher_fence = Q3 + 1.5 * IQR
```
