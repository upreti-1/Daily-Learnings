### Done to remove biasness form the data. Techniques are Upsampling and Downsampling.

```python
# 1. UpSampling & DownSampling

import numpy as np
import pandas as pd

# set the random seed for reporducibility
np.random.seed(123)


# creating a dataframe with two classes
n_samples = 1000
class_0_ratio = 0.9
n_class_0 = int(n_samples * class_0_ratio)
n_class_1 = n_samples - n_class_0
```

```python
n_class_0, n_class_1
```

```python
# creating a imbalance dataset

class_0 = pd.DataFrame({
    'feature_1' : np.random.normal(loc=0, scale = 1, size = n_class_0),
    'feature_2' : np.random.normal(loc=0, scale = 1, size = n_class_0),
    'target' : [0] * n_class_0
    
})

class_1 = pd.DataFrame({
    'feature_1' : np.random.normal(loc=2, scale = 1, size = n_class_1),
    'feature_2' : np.random.normal(loc=2, scale = 1, size = n_class_1),
    'target' : [1] * n_class_1
    
})
```

```python
df = pd.concat([class_0, class_1]).reset_index(drop = True)
```

```python
df.head()
```

```python
df.tail()
```

```python
df['target'].value_counts()
```

```python
# using upsampling now
df_minority = df[df['target'] == 1]
df_majority = df[df['target'] == 0]
```

```python
from sklearn.utils import resample
df_minority_upsampling = resample(df_minority, replace = True #samples with replacement
         , n_samples = len(df_majority)
         ,random_state = 42)
```

```python
df_minority_upsampling.shape
```

```python
df_minority_upsampling.head()
```

```python
df_upsampled = pd.concat([df_majority, df_minority_upsampling])
```

```python
df_upsampled['target'].value_counts()
```

```python
# DownSampling now
# creating a imbalance dataset

class_0 = pd.DataFrame({
    'feature_1' : np.random.normal(loc=0, scale = 1, size = n_class_0),
    'feature_2' : np.random.normal(loc=0, scale = 1, size = n_class_0),
    'target' : [0] * n_class_0
    
})

class_1 = pd.DataFrame({
    'feature_1' : np.random.normal(loc=2, scale = 1, size = n_class_1),
    'feature_2' : np.random.normal(loc=2, scale = 1, size = n_class_1),
    'target' : [1] * n_class_1
    
})
```

```python
df = pd.concat([class_0, class_1]).reset_index(drop = True)
```

```python
df['target'].value_counts()
```

```python
# using downsampling now
df_minority = df[df['target'] == 1]
df_majority = df[df['target'] == 0]
```

```python
from sklearn.utils import resample
df_majority_downsampling = resample(df_majority, replace = False  # because we have to remove the datapoints and not add it
         , n_samples = len(df_minority)
         ,random_state = 42)
```

```python
df_majority_downsampling.shape
```

```python
df_downsampled = pd.concat([df_minority,df_majority_downsampling])
```

```python
df_downsampled.target.value_counts()
```
