```python
import seaborn as sns
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
data = sns.load_dataset('tips')
```

```python
data.head()
```

```python
df = pd.DataFrame(data)
```

```python
encoder = OneHotEncoder()
```

```python
encoded1 = encoder.fit_transform(df[['sex']]).toarray()
```

```python
encoded2 = encoder.fit_transform(df[['smoker']]).toarray()
```

```python
sex_encoded = pd.DataFrame(encoded1, columns = encoder.get_feature_names_out())
smoker_encoded = pd.DataFrame(encoded2, columns = encoder.get_feature_names_out())
```

```python
pd.concat([sex_encoded, smoker_encoded], axis = 1)
```
