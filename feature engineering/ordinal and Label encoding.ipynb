# Label Encoding

### Assigning a unique numerical label to each category in a variable.

```python
import pandas as pd
```

```python
# create a simple DataFrame
df = pd.DataFrame({
    'color' : ['red','blue','green','yellow','red','blue']
})
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
lbl_encoder = LabelEncoder()
```

```python
lbl_encoder.fit_transform(df[['color']])
```

```python
lbl_encoder.transform([['green']])
```

```python
lbl_encoder.transform([['red']])
```

# Ordinal Encoding

### used to encode categorical data that have an intrinsic order or ranking. In this technique, each category is assigned a numerical value bases on its position in the order. It assigns ranks to the features.

```python
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder
```

```python
df = pd.DataFrame({
    'size': ['small', 'medium', 'large', 'small', 'large']
})
```

```python
df
```

```python
# assigning ranking now
encoder = OrdinalEncoder(categories=[['small', 'medium', 'large']])
encoder.fit_transform(df[['size']])
```

```python
encoder.transform([['small']])
```
