# SVM Kernals Indepth Intution And Practical Explaination

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-5.0, 5.0, 100)
y = np.sqrt(10**2 - x**2)
y = np.hstack([y, -y])
x = np.hstack([x, -x])
```

```python
plt.scatter(y, x)
```

```python
x1 = np.linspace(-5.0, 5.0, 100)
y1 = np.sqrt(5**2 - x1**2)
y1 = np.hstack([y1, -y1])
x1 = np.hstack([x1, -x1])
```

```python
plt.scatter(y, x)
plt.scatter(y1, x1)
```

```python
x
```

```python
df1 = pd.DataFrame(np.vstack([y,x]).T, columns = ['X1', 'X2'])
df1['Y'] = 0
df2 = pd.DataFrame(np.vstack([y1,x1]).T, columns = ['X1', 'X2'])
df2['Y'] = 1
df = pd.concat([df1, df2])
df.head(3)
```

```python
# Independent and Dependent features
X = df.iloc[:, :2]
y = df.Y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.25, random_state= 42)
```

```python
y_train
```

# Polynimial Kernal


```python
# we need to find components for the polynimial components
# X1, X2, X1_square, X2_square, X1*X2
df['X1_square'] = df['X1']**2
df['X2_square'] = df['X2']**2
df['X1*X2'] = (df['X1'] * df['X2'])
df.head()
```

```python
# dependent and independent features
X = df[['X1', 'X2', 'X1_square', 'X2_square', 'X1*X2']]
y = df['Y']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.25, random_state= 42)
```

```python
import plotly.express as px
fig = px.scatter_3d(df, x = 'X1', y = 'X2', z = 'X1*X2', color = 'Y')
fig.show()
```

```python
import plotly.express as px
fig = px.scatter_3d(df, x = 'X1_square', y = 'X2_square', z = 'X1*X2', color = 'Y')
fig.show()
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
classifier = SVC(kernel = 'linear')
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
# This kernal directly created X1_squared X2_squared and everything automatically. No manual labour
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
classifier = SVC(kernel = 'poly')
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
# This kernal directly created X1_squared X2_squared and everything automatically. No manual labour
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
classifier = SVC(kernel = 'rbf')
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)
accuracy_score(y_test, y_pred)
```
