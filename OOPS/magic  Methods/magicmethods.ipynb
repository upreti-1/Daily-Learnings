### Basic Magic Methods

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

person = Person('prashant', 22)
print (person)
```

```python
# overriding now
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __str__(self):
        return f"{self.name}, {self.age} years old"

person = Person('prashant', 22)
print (person)

# yaha __str__ hamle pailai modify garera use gardeu so tala use vayena tyo
```

```python
# overriding now
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __str__(self):      # yaha __str__ hamle pailai modify garera use gardeu so tala use vayena tyo
        return f"{self.name}, {self.age} years old"

    def __repr__(self):
        return f"name = {self.name}, age = {self.age} years old"
person = Person('prashant', 22)
print (person)
print (repr(person))
```
