# Abstraction

## Concept of hiding complex implementation details and sharing only the necessary features of an object. This helps in reducing programming complexity and effort.

```python
from abc import ABC, abstractmethod

# Abstract class
class Vehicle(ABC):
    def drive(self):
        print('The vechicle is used for driving')

    @abstractmethod  # says "their is one functionality available for you, paxi implement this function by your own way"
    def start_engine(self):
        pass

class Car(Vehicle):
    def start_engine(self):
        print ('Car engine started')

def operate_Vehicle(vehicle):
    vehicle.start_engine()
    vehicle.drive()

car = Car()

operate_Vehicle(car)
```
