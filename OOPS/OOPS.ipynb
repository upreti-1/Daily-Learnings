```python
# Instance variables and methods
class Dog:
    # constructor
    def __init__(self, name, age): # name and age are attributes of the class Dog
        self.name = name    # self.name is the instance variable of the class Dog
        self.age = age      # yo pani instance variable
```

```python
dog1 = Dog('tommy', 10)
print(dog1)
print(dog1.name, dog1.age)
```

```python
dog2 = Dog('buddy', 22)
print(dog2.name, dog2.age)
```

```python
# class with instance method
class Cat:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def bark(self): # this is an instance method
        print(f'{self.name} says meow')
        print("meow")
```

```python
cat1 = Cat('billo', 2)
cat1.bark()             # calling the instance method in class Cat

cat2 = Cat('lusy', 1)
cat2.bark()
```

```python
# modelling a bank account

# Define a class bank account
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = balance

    def deposit (self, amount):
        self.balance += amount
        print(f'Your updated balance is {self.balance}+{amount} = {self.balance+amount}')
    
    def withdrawl (self, amount):
        if (amount>self.balance):
            print(f"The requested amount is higher then your balance : )")
            return 0
        else:
            print(f'Withdrawl sucessful, Your new balance is : {self.balance} - {amount} = {self.balance - amount}')
    
    def get_balance(self):
        print(f'Remaining amount is : {self.balance}')


# creating an object

account1 = BankAccount('Prashant', 200)
print(account1.balance)

account1.deposit(100)


account1.get_balance()

account1.withdrawl(100)
```
