# Inheritance in OOP allows a class to inherit attibutes and methods form another class.

## single inheritance and Multiple inheritance 

### single inheritance

```python
# parent class
class Car:
    def __init__(self, windows, doors, engine):
        self.windows = windows
        self.doors = doors
        self.engine = engine

    def drive(self):
        print(f"The person will drive {self.engine} car")
```

```python
# This is a object
car1 = Car(4,2,'petrol')
car1.drive()
```

```python
# Tesla is the child class inheriting form the parent class
class Tesla(Car):
    def __init__(self, windows, doors, engine, is_self_driving):
        super().__init__(windows, doors, engine)  # coming the attrubutes form the parent class
        self.is_self_driving = is_self_driving

    def self_driving(self):
        print(f"Tesla supports self dring: {self.is_self_driving}")
```

```python
tesla1 = Tesla(4,5,'electric', True)
tesla1.self_driving()
```

```python
tesla1.drive()
```

```python
tesla2 = Tesla(2,3,'diesel', False)
tesla2.self_driving()
```

```python
tesla2.drive()
```

## Another example

```python
class man():
    def __init__(self, name, age, height):
        self.name = name
        self.age = age
        self.height = height

    def whatage(self):
        print(f"The age of {self.name} is {self.age}")
```

```python
class boy(man):
    def __init__(self, name, age, height, year):
        super().__init__(name, age, height)
        self.year = year

    def say_year(self):
        print(f'The year is {self.year}')

    def details(self):
        print(f'Name : {self.name}, age = {self.height}, height = {self.height}, born in = {self.year}')
```

```python
boy1 = boy('sparsh', 12, 4, 2000)
boy1.say_year()
```

```python
boy1.details()
```

```python
boy1.whatage()
```

# Multiple inheritance

```python
# when a class inheritances form more then one class
```

```python
# base classes 1
class animal():
    def __init__(self, name):
        self.name = name
    
    def speak(self):
        print(f'subclasses must implement this classes')
```

```python
# base classes 2
class pet():
    def __init__(self, owner):
        self.owner = owner
    
    def owner_name (self):
        print(f'The owner is : {self.owner}')
```

```python
# derived class
class Dog(animal, pet):
    def __init__(self, name, owner):
        animal.__init__(self, name)
        pet.__init__(self, owner)

    def speak(self):
        return f'{self.name} says bhau'
```

```python
dog1 = Dog('tommy', 'prashant')
dog1.speak()
```

```python
# multilevel inheritance
class breed(Dog):
    def __init__(self, name, owner, breed_type):
        super().__init__(name, owner)
        self.breed_type = breed_type

    def say_details(self):
        return f'name : {self.name}, owner = {self.owner}, breed is = {self.breed_type}'
```

```python
kukur = breed('toni', 'sparsh', 'german sepherd')
kukur.say_details()
```
