# Encapsulation And Abstraction

### These are two fundamental principles of OOPS that help in designing robust, maintainable, & reusable code. Encapsulation involves bundling data and methods that operate on the data within a single unit, while abstraction involves hiding hiding complex implementation details and exposing only the necessary features.

## Encapsulation

### Concept of wrapping data (variables) and methods (functions) together as a single unit. It restricts direct access of some of the object's components, which is means of preventing accidental inference and misuse of the data.

```python
class Person:
    def __init__(self, name, age, gender):
        self._name = name   # protected variable
        self._age = age     # protected variable
        self.gender = gender # public variable

# cannot use it outside the class but we can access it form derived class

class Employee(Person):
    def __init__(self, name, age, gender, salary):
        super().__init__(name, age, gender)
        self.salary = salary

employee1 = Employee('Prashant', 20, 'Male', 300000)
print(employee1._name)

print(employee1.salary)



class man:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def get_gender(self, person):
        return person.gender
    
    def get_name(self, employe):
        return employe._name


man1 = man('hello', 20)
print(man1.get_gender(employee1))
print(man1.get_name(employee1))
```

```python
# Encapsulation with Getter and Setter
# basically Getter le dekhauxa and Setter le change garxa in the same class without exposing the core private variables outside the class.
# we can access the variables in the class without disturbing the underlying core structure outside the class.
class Person:
    def __init__(self, name, age):
        self.__name = name  # Private access modifier or variable
        self.__age = age  # Private access modifier or variable

    # Getter method for name
    def get_name(self):
        return self.__name
        
    # setter method for name
    def set_name(self, name):
        self.__name = name

    # Getter method for age
    def get_age (self):
        return self.__age
        
    # Setter method for age
    def set_age(self, age):
        if age > 0:
            self.__age = age
        else:
            print ('Age cannot be less then 0')

# Access and modify private variables using Getter and Setter
person0 = Person('Prashant', 20)
print(person0.get_name())
print(person0.get_age())
person0.set_name('Akash')
person0.set_age(23)
print (person0.get_name())
print (person0.get_age())
# yaha halme class ko bahirai bata private variables access and use gardai xau using Getter and Setter
```
