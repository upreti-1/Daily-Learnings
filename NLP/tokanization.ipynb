# TOKANIZATION.

```python
import nltk
# This downloads the most common tokenizers and corpora
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
```

```python
corpus = """Hello Welcome, Today's the day. we are gonna learn about NLP.
Lets watch the entire course! To become good at NLP.
Lets do it!
"""
```

```python
print(corpus)
```

```python
# Tokanization
# Changing sentence into paragraph
from nltk.tokenize import sent_tokenize
documents = sent_tokenize(corpus)
```

```python
# line tokenize: it make tokens of different lines
from nltk.tokenize import line_tokenize
line_tokenize(corpus)
```

```python
type(documents)
```

```python
for sentence in documents:
    print (sentence)
```

```python
# Tokenization
# Paragraph --> Words and Sentences --> Words
from nltk.tokenize import word_tokenize
word_tokenize(corpus)
```

```python
for sentence in documents:
    print(word_tokenize(sentence))
```

```python
from nltk.tokenize import wordpunct_tokenize
wordpunct_tokenize(corpus)
```

```python
from nltk.tokenize import TreebankWordTokenizer
tokenizer = TreebankWordTokenizer()
done = tokenizer.tokenize(corpus)
print(done)
```

```python
from nltk.tokenize import TreebankWordDetokenizer
detokenizer = TreebankWordDetokenizer()
undone = detokenizer.detokenize(done)
print(undone)
```
