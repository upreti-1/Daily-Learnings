# Stemming and its Types

```python
words = ['eating', 'eats', 'eaten', 'writing', 'writes', 'programming', 'programs', 'histroy', "There's"]
```

### PorterStemmer

```python
from nltk.stem import PorterStemmer
```

```python
stemming = PorterStemmer()
for word in words:
    print(word + ' ---> ' + stemming.stem(word))
```

```python
stemming.stem('congratulations')    # disadv: Lost of meaning in some cases. We use Lematization
```

# RegexStemmer class
We can easily implement Regular Expression Stemmer algo. using this. It takes a single regular expression and removes any prefix and suffix that matches the expression.

```python
from nltk.stem import RegexpStemmer
reg_stemmer = RegexpStemmer(regexp= 'ing$|s$|e$|able$', min=4)
```

```python
reg_stemmer.stem('dog')
```

# Snowball Stemmer

```python
from nltk.stem import SnowballStemmer
snowball_stemmer = SnowballStemmer('english')
```

```python
for word in words:
    print(word + ' ---> ' + snowball_stemmer.stem(word))
```
