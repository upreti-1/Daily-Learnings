### Data cleaning

```python
df['Reviews'].unique()
```

```python
df['Reviews'].astype(int)
```

```python
df['Reviews'].str.isnumeric().sum()
```

```python
df[~df['Reviews'].str.isnumeric()]
```

```python
df_copy = df.copy()
```

```python
df_copy=df_copy.drop(df_copy.index[10472])
```

```python
df_copy[~df_copy['Reviews'].str.isnumeric()]
```

```python
df_copy['Reviews'] = df_copy['Reviews'].astype(int)
```

```python
df_copy.info()
```

```python
df_copy.head(2)
```

```python
df_copy['Size'].unique()
```

```python
df_copy['Size'] = df_copy['Size'].str.replace('M', '000')
```

```python
df_copy['Size'].isnull().sum()
```

```python
df_copy['Size'] = df_copy['Size'].str.replace('k', '')
```

```python
df_copy['Size'].unique()
```

```python
df_copy['Size'] = df_copy['Size'].replace('Varies with device', np.nan)
```

```python
df_copy['Size'] = df_copy['Size'].astype(float)
```

```python
df_copy.info()
```

```python
df_copy['Size'].isnull().sum()
```

```python
df_copy.head()
```

```python
df_copy['Installs'].unique()
```

```python
df_copy['Price'].unique()
```

```python
chars_to_remove = ['+', '$', ',']
cloumns_to_clean = ['Installs', 'Price']

for items in chars_to_remove:
    for cols in cloumns_to_clean:
        df_copy[cols] = df_copy[cols].str.replace(items, '')
```

```python
df_copy['Price'].unique()
```

```python
df_copy['Installs'].unique()
```

```python
df_copy.head(2)
```

```python
df_copy["Installs"] = df_copy["Installs"].astype(int)
```

```python
df_copy["Price"] = df_copy["Price"].astype(float)
```

```python
df_copy.info()
```

```python
df_copy['Last Updated'].unique()
```

```python
df_copy['Last Updated'] = pd.to_datetime(df_copy['Last Updated'])
```

```python
df_copy['Last Updated']
```

```python
df_copy['Day'] = df_copy['Last Updated'].dt.day
df_copy['Month'] = df_copy['Last Updated'].dt.month
df_copy['year'] = df_copy['Last Updated'].dt.year
```

```python
df_copy.info()
```

```python
df_copy.head()
```

```python
df_copy.to_csv('data/google_cleaned.csv')
```

## Explore data more

```python
df_copy.info()
```

```python
numerical_features = [feature for feature in df_copy.columns if df_copy[feature].dtype != 'O']
categorical_features = [feature for feature in df_copy.columns if df_copy[feature].dtype == 'O']

print('We have {} numerical features: {}'.format(len(numerical_features), numerical_features))
print('We have {} numerical features: {}'.format(len(categorical_features), categorical_features))

```

```python
### proportion of count data on categorial column
for col in categorical_features:
    print(df_copy[col].value_counts(normalize=True)*100)
    print('---------------------------')
```

```python
## proportion of count data on numerical columns
plt.figure(figsize=(15,15))
plt.suptitle('Univariate Analysis of Numerical Features', fontsize = 20)

for i in range (0, len(numerical_features)):
    plt.subplot(5,3,i+1)
    sns.kdeplot(x = df_copy[numerical_features[i]], shade = True, color = 'r')
    plt.xlabel(numerical_features[i])
    plt.tight_layout()
```

### observation 
- Rating and year is left skewed while reviews, size, installs and price are right skewed

```python
## proportion of count data on categorical columns
plt.figure(figsize=(15,15))
plt.suptitle('Univariate Analysis of Categorical Features', fontsize = 20)
category = ['Type', 'Content Rating']

for i in range (0, len(category)):
    plt.subplot(2,2,i+1)
    sns.countplot(x = df_copy[category[i]], palette='Set2')
    plt.xlabel(category[i])
    plt.xticks(rotation = 45)
    plt.tight_layout()
```
