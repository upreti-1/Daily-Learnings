# Pandas - DataFrame and Series

### Powerful data manupulation library in Python, widely used for data analusis and data cleaning. It provides two primary data structures: Series and DataFrame. A series is one-dimensional array-like object, while Dataframe is two-dimensional, size-mutable, and potentially heterogeneous tabular data structure with labelled axes (rows and columns)

### Series

```python

# 1-D array like object that can hold any data type. It is similar to column in a table.
# creating series from list
import pandas as pd
data = [1,2,3,4,5]
series = pd.Series(data)
print(series)
print(type(series))
```

```python
# creating series from dictionary
data = {'a':1, 'b':2, 'c':3}
series_dict = pd.Series(data)
print(series_dict)
print(type(series_dict))
```

```python
# index manupulation in Series
data = [1,3,5,7]
index = ['a','c','e', 'g']
pd.Series(data, index = index)
```

### DataFrame

```python
# 2-D similar to a table in database
## creating a DataFrame form a dictionary of list

data = {
    'name': ['prashant', 'john', 'jack'],
    'age' : [23,44,35],
    'city' : ['bangalore','delhi','patna']
}

df = pd.DataFrame(data)
print(df)
print('\n')
print(type(df))
```

```python
import numpy as np
print(np.array(data))
```

```python
# create a data frame from a list of dictionaries

data = [
    {'name':'prashant', 'age': 32, 'city': 'Bangalore'},
    {'name':'john', 'age': 32, 'city': 'Bangalore'},
    {'name':'jack', 'age': 32, 'city': 'Bangalore'},
    {'name':'khoi', 'age': 32, 'city': 'Bangalore'},
]

df = pd.DataFrame(data)
print(df)
```
