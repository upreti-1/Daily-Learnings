# Python Memory Management

### Memory management in python involves a combination of automatic garbage collection, reference counting and various internal optimization to efficient manage memory allocation and deallocation

## Reference Counting

### It is primary method Python uses to manage memory. Each object in Python maintains a count of references pointing to it. When the reference count drops to 0, the memory occupied by the object is deallocated.

```python
import sys  # system configuration

a = []
print(sys.getrefcount(a))  # O/P : 2 because one reference is coming form 'a' variable and another form getrefcount() function
```

```python
b = a
print (sys.getrefcount(b))
```

```python
print (sys.getrefcount(a))
```

```python
del b
```

```python
print (sys.getrefcount(a))
```

## Garbage Collection

### Python includes cyclic garbage collector to handle reference cycles. Reference cycles occur when reference each other, preventing their reference counts form reaching zero.

```python
import gc   # garbage collector

gc.enable()  # enabling garbage collection
gc.disable()   # to disable it
```

```python
gc.collect()   # these are values are present and unreachable
```

```python
# get garbage collection stats
print(gc.get_stats())
```

```python
# get unrachable objects
print(gc.garbage)
```

## Memory management best practices

### 1. use local variables
### 2. Avoid circular references (e.g. : a = b and b = a): can lead to memory leaks 
### 3. Use generators
### 4. Explicitly delete objects
### 5. Profile Memory Usage : tools like tracemalloc and memory_profiler

```python
# handling circular references with manually trigerring garbage collection

import gc

class MyObject:
    def __init__(self, name):
        self.name = name
        print(f'Object {self.name} created')

    def __del__(self):
        print(f'Object {self.name} deleted')

# creating circular reference
obj1 = MyObject('obj1')
obj2 = MyObject('obj2')
obj1.ref = obj2         # bad practices
obj2.ref = obj1         # bad practices

del obj1
del obj2

# deleting objects may delete the objects but their references does not gets deleted and is stored. To delete it, we manually collect the references in Garbage Collector
gc.collect()         # O/P shows deletion only because of this line as the previous del ObjName function could not delete it because of circular reference.
```

```python
# Generators for memory efficiency

def generate_numbers(n):
    for i in range (n):
        yield i

for num in generate_numbers(10000):
    print(num)
    if num > 10:
        break
```

### Profiling memory usage

```python
# using tracemalloc

import tracemalloc

def create_list():
    return [i for i in range (1000)]

def main():
    tracemalloc.start()

    create_list()

    snapshot = tracemalloc.take_snapshot()
    top_stats = snapshot.statistics('lineno')

    print('Memory management')
    for stat in top_stats[::]:
        print(stat)
```

```python
main()
```
