### Reading a whole file

```python
# Reading a Whole File
with open('source.txt', 'r') as file:
    content = file.read()
    print(content)
```

```python
# reading line by line
with open('source.txt','r') as file:

    for line1 in file:
        print(line1.strip())
```

```python
# Overwriting a File
with open('destination.txt','w') as file:
    file.write('1221')
    file.write('This is a new file')

with open('destination.txt', 'r') as file1:
    content = file1.read()
    print(content)
```

```python
# appending a file
with open('example.txt', 'a') as file:
    file.write('This is a new appended line')

with open('example.txt','r') as file:
    for line in file:
        print(line)
```

```python
# to write multiple lines
lines = ['\nfirst\n','second\n','third\n']
with open('example.txt','a') as file:
    file.writelines(lines)
    
with open('example.txt','r') as file:
    for line in file:
        print(line)
```

```python
with open('example.txt','w+') as file:
    file.write('Hello world\n')
    file.write('Hello world1\n')

    # now to read, move the cursor to the top of the file
    file.seek(0)

    # read the content the file
    content = file.read()
    print(content)
```

## Working with file path

```python
# create a new directory
import os
os.mkdir('package')
print(f'Directory has been created')
```

```python
# list all the files and directoreis in an directory
os.listdir('.')
```

```python
# join path
folder_name = 'folder'
file_name = 'file.txt'
full_path = os.path.join(folder_name, file_name)
print(full_path)
```

```python
# join path
folder_name = 'folder'
file_name = 'file.txt'
full_path = os.path.join(os.getcwd(),folder_name, file_name)
print(full_path)
```

```python
relative_path = 'example.txt'
absolute_path = os.path.abspath(relative_path)
print(absolute_path)
```
