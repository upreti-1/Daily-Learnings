```python
# Exception Handling

try:
    a = b
except:
    print("The variable has not been assigned")
```

```python
try:
    n = int(input("enter a number  : "))
    result = n-2
    print(result)
except ValueError as ve:
    print(ve)
    print('enter a correct numerical value : ) ')
except Exception as ex:
    print (ex)
    
```

```python
# try, except, else block
# else block will execute if there is no exception
try:
    n = int(input("enter a number  : "))
    result = n-2
    print(result)
except ValueError as ve:
    print(ve)
    print('enter a correct numerical value : ) ')
except Exception as ex:
    print (ex)
else:
    print('Result is calculated')
```

```python
# try, except, else, finally block
try:
    n = int(input("enter a number  : "))
    result = n-2
    print(result)
except ValueError as ve:
    print(ve)
    print('enter a correct numerical value : ) ')
except Exception as ex:
    print (ex)
else:
    print('Result is calculated')
finally:
    print('\nThank You!')
```
