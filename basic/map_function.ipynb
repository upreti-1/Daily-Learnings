## MAP : It applies a given function to all the items in a given list

```python
def square(x):
    return x*x
square(5)
```

```python
numbers = [1,2,3,4,5]
list(map(square, numbers))
```

### lambda function with map

```python
numbers = [1, 2, 3, 4, 5]
list(map(lambda x:x*x, numbers))  # Lambda
```

### Mapping multiple Iterables

```python
numbers1 = [1, 2, 3]
numbers2 = [4, 5, 6]
list(map(lambda x,y: x+y, numbers1, numbers2))
```

### Map() to convert list of strings to integer

#### Typecasting

```python
numbers = ['1', '2', '3', '4', '5']
list(map(int, numbers))
```
